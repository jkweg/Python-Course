## **Teoria. Fabryka konfiguracji (lab.5)**

**Wzorzec projektowy Factory**

Wzorzec projektowy Factory (fabryka) należy do wzorców kreacyjnych i służy do kontrolowania procesu tworzenia obiektów. Jego główną ideą jest oddzielenie logiki tworzenia obiektów od ich użycia.

Zamiast tworzyć obiekty bezpośrednio w różnych miejscach programu, wprowadza się specjalny mechanizm – fabrykę – która odpowiada za ich instancjonowanie. Fabryka na podstawie przekazanych danych (np. typu, nazwy, konfiguracji) decyduje, jaki obiekt należy utworzyć.

Takie podejście zwiększa czytelność kodu oraz ułatwia jego rozwój. W przypadku zmiany sposobu tworzenia obiektów lub dodania nowych typów, modyfikacje są ograniczone do jednego miejsca. Wzorzec ten wspiera zasadę pojedynczej odpowiedzialności oraz zmniejsza zależności między komponentami systemu.

Factory jest szczególnie użyteczny w sytuacjach, gdy:

* istnieje wiele możliwych typów obiektów,
* wybór konkretnego typu następuje w trakcie działania programu,
* tworzenie obiektów jest złożone lub zależne od danych wejściowych.



Operator ```**data ```

Operator ** w Pythonie służy do rozpakowywania słownika do argumentów nazwanych funkcji lub konstruktora klasy.

Jeżeli słownik zawiera pary klucz–wartość, operator ** pozwala przekazać je jako parametry, gdzie klucze odpowiadają nazwom argumentów, a wartości – ich wartościom.

Warunkiem poprawnego działania jest zgodność nazw kluczy w słowniku z nazwami parametrów funkcji lub pól klasy. Dzięki temu możliwe jest dynamiczne tworzenie obiektów na podstawie danych, np. wczytanych z pliku konfiguracyjnego.

Mechanizm ten znacząco upraszcza kod i eliminuje konieczność ręcznego przypisywania każdej wartości, co jest szczególnie istotne przy pracy z danymi konfiguracyjnymi.

**Enum i StrEnum**

Enum (enumeration) to typ danych służący do definiowania zbioru stałych wartości. Każdy element enuma ma unikalną nazwę i przypisaną wartość.

Zastosowanie enumów pozwala zastąpić „surowe” wartości (np. napisy lub liczby) bardziej czytelnymi i bezpiecznymi konstrukcjami. Dzięki temu ogranicza się ryzyko błędów, takich jak literówki, oraz poprawia się czytelność kodu.

StrEnum jest szczególnym przypadkiem enuma, w którym wartości są typu tekstowego (string). Łączy on zalety enumów z wygodą pracy na napisach.

Zastosowanie enumów jest szczególnie korzystne w sytuacjach, gdy:

* istnieje ograniczony zbiór dopuszczalnych wartości,
* wartości te mają znaczenie semantyczne,
* ważna jest jednoznaczność i bezpieczeństwo typów.

**Konstrukcja match-case**

Konstrukcja match-case w Pythonie jest mechanizmem dopasowania wzorców (pattern matching), który pozwala wykonywać różne działania w zależności od wartości zmiennej.

Jest to bardziej rozbudowana i czytelna alternatywa dla wielokrotnych instrukcji warunkowych typu if-elif-else. Pozwala ona w przejrzysty sposób obsłużyć wiele przypadków, szczególnie gdy decyzja zależy od jednej zmiennej.

match-case umożliwia:

* porównywanie wartości,
* dopasowanie do konkretnych typów,
* pracę z bardziej złożonymi strukturami danych.



**Klasa ApplicationConfig**

Klasa ApplicationConfig pełni rolę obiektu agregującego konfigurację całej aplikacji. Jej zadaniem jest przechowywanie i zarządzanie wszystkimi sekcjami konfiguracji w jednym miejscu.

Zamiast operować na rozproszonych danych (np. wielu słownikach), cała konfiguracja jest reprezentowana jako jeden spójny obiekt. Poszczególne elementy konfiguracji są przechowywane jako pola lub kolekcje obiektów.

Takie podejście umożliwia:

* centralne zarządzanie konfiguracją,
* łatwiejszą walidację wszystkich danych,
* uporządkowaną strukturę programu,
* wygodne przekazywanie konfiguracji między komponentami systemu.

Klasa ta często udostępnia metody umożliwiające operacje globalne, takie jak walidacja całej konfiguracji lub jej prezentacja.

**Rejestr fabryk**

Rejestr fabryk to mechanizm rozszerzający wzorzec Factory, który umożliwia dynamiczne zarządzanie typami obiektów, jakie mogą być tworzone.

Zamiast stosować na stałe zdefiniowaną logikę wyboru obiektów (np. instrukcje warunkowe), wykorzystuje się strukturę danych (np. słownik), która mapuje identyfikatory na odpowiednie klasy lub funkcje tworzące obiekty.

Dzięki temu:

* można łatwo dodawać nowe typy obiektów bez modyfikacji istniejącego kodu,
* system staje się bardziej elastyczny i rozszerzalny,
* zmniejsza się liczba zależności między komponentami.

Rejestr fabryk jest szczególnie przydatny w systemach modularnych oraz takich, które są rozwijane etapowo. Umożliwia on rozszerzanie funkcjonalności poprzez rejestrację nowych elementów, bez ingerencji w istniejącą logikę działania.
